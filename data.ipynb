```python
import pandas as pd
import seaborn as sns
import matplotlib as plt

df = pd.read_csv("classified_data.csv")
df
```

```python
## remove row if is_generative or generative_type is NaN
df = df.dropna(subset=['is_generative', 'generative_type'])
df
```

```python
num_generative = df['is_generative'].sum()
total_entries = len(df)
percentage_generative = (num_generative / total_entries) * 100
print(f"Percentage of generative entries: {percentage_generative:.2f}%")
```

```python
## remove row if is_generative is False
df = df[df['is_generative'] == True]
df
```

```python
## names should be split by | and each name should have its own row
df_names = df.assign(Name=df['Name'].str.split('|')).explode('Name')
df_names.groupby('Name').size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```

```python
df.groupby('generative_type').size().reset_index(name='counts')
plt.pyplot.figure(figsize=(10,6))
sns.countplot(data=df, x='generative_type', order=df['generative_type'].value_counts().index)
plt.pyplot.xticks(rotation=45)
plt.pyplot.title('Distribution of Generative Types')
plt.pyplot.show()
```

```python
## count per Organisational Unit Name
## separate by | and each name should have its own row
df_org = df.assign(**{'Organisational unit name': df['Organisational unit name'].str.split('|')}).explode('Organisational unit name')
df_org['Organisational unit name'] = df_org['Organisational unit name'].str.strip()
df_org.groupby('Organisational unit name').size().reset_index(name='counts').sort_values(by='counts', ascending=False)
```
